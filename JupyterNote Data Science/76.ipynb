```python
import numpy as np

for i in df :
    a = df[i]=='?'
    print(a.value_counts())
#pekerjaan,kelas pekerja
```

```python
df['Pekerjaan'] = df['Pekerjaan'].replace({'?': np.nan})
```

```python
df['Kelas Pekerja'] = df['Kelas Pekerja'].replace({'?': np.nan})
```

```python
df['Pekerjaan'] = df['Pekerjaan'].fillna('Spesialis')
```

```python
df['Pekerjaan'].value_counts()
```

```python
df['Kelas Pekerja'] = df['Kelas Pekerja'].fillna('Wiraswasta')
```

```python
df['Kelas Pekerja'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Kelas Pekerja'].value_counts()
pekerjaan = {
    'Pemerintah Negara':7,
    'Pemerintah Provinsi': 6,
    'Pemerintah Lokal': 5,
    'Wiraswasta':4,
    'Pekerja Bebas Perusahaan': 3,
    'Pekerja Bebas Bukan Perusahan':2,
    'Tanpa di Bayar' : 1,
    'Tidak Pernah Bekerja':0
}
# pemerintah negara>provinsi>lokal>wiraswasta>pekerja bebas perusahaan>pekerja bebas bkn perusahaan>tdk dibayar> tdk pernah bekerja
```

```python
df['Kelas Pekerja'] = df['Kelas Pekerja'].replace(pekerjaan)
df['Kelas Pekerja'].value_counts()
```

```python
df['Pendidikan'].value_counts()
# doktor > master > sarjana > d4 > D3 > Pendidikan Tinggi > Sekolah Professional > SMA > 12th > 11th > 10th
# > 9th > 7th-8th > SD > 5th-6th > 1st-4th
pendidikan = {
    'Doktor':9,
    'Master': 9,
    'Sarjana': 9,
    'D4':9,
    'D3':9,
    'Pendidikan Tinggi':9,
    'Sekolah Professional':8,
    'SMA':7,
    '12th':7,
    '11th':6,
    '10th':5,
    '9th':4,
    '7th-8th':3,
    'SD':2,
    '5th-6th':1,
    '1st-4th':0
}
df['Pendidikan'] = df['Pendidikan'].replace(pendidikan)
```

```python
df.describe()
```

```python
df['Pekerjaan'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Status Perkawinan','Jenis Kelamin','Pekerjaan'])
```

```python
df = df.drop(columns=['id','Berat Akhir'])
```

```python
df
```

```python
data = df.drop(columns=['Gaji'])
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
dataNorm = std.fit_transform(data)

dfNorm = pd.DataFrame(dataNorm, columns=['Umur','Kelas Pekerja',
                                           'Pendidikan','Jmlh Tahun Pendidikan','Keuntungan Kapital','Kerugian Capital',
                                           'Jam per Minggu','Status Perkawinan_Belum Pernah Menikah',
                                           'Status Perkawinan_Berpisah','Status Perkawinan_Cerai','Status Perkawinan_Janda',
                                           'Status Perkawinan_Menikah','Status Perkawinan_Menikah LDR',
                                           'Jenis Kelamin_Laki2','Jenis Kelamin_Perempuan','Pekerjaan_Asisten Rumah Tangga','Pekerjaan_Ekesekutif Managerial','Pekerjaan_Mesin Inspeksi','Pekerjaan_Pembersih',
                                         'Pekerjaan_Pemuka Agama',
                                         'Pekerjaan_Penjaga','Pekerjaan_Perbaikan Kerajinan','Pekerjaan_Petani','Pekerjaan_Sales','Pekerjaan_Servis Lainnya',
                                         'Pekerjaan_Spesialis','Pekerjaan_Supir','Pekerjaan_Tech-support','Pekerjaan_Tentara'])
dfNorm
```

```python
X = dfNorm
y = df['Gaji']
```

```python
y.value_counts()/df.shape[0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=21, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

model = RandomForestClassifier()
#params = {'n_neighbors': np.arange(3,11), 'weights':['distance', 'uniform']}
param = {'criterion':['gini', 'entropy'], 'random_state':np.arange(10)}
gscv = GridSearchCV(model, param_grid=param, scoring='roc_auc', cv=10)
gscv.fit(X_train,y_train)
gscv.best_params_
```

```python
gscv.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini',random_state= 4)
model.fit(X,y)
```

```python
y_new = model.predict(X_test)
```

```python
y_new
```

```python
y_new = model.predict_proba(X_test)
```

```python
y_pred = y_new[:, 1]
```
