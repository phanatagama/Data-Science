```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales (1).csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
rata = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
rata
```

```python
rata2 = rata.transpose()
rata2
```

Expected Output :



<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
fig, ax = plt.subplots(figsize=(16,8))
x = np.arange(0,8,2)
y = rata2['Action']
y1 = rata2['Adventure']
y2 = rata2['Fighting']
y3 = rata2['Misc']
y4 = rata2['Platform']
y5 = rata2['Puzzle']
y6 = rata2['Racing']
y7 = rata2['Role-Playing']
y8 = rata2['Shooter']
y9 = rata2['Simulation']
y10 = rata2['Sports']
y11 = rata2['Strategy']

ax.bar(x-0.5,y, width=0.1, label='Action')
ax.bar(x-0.4,y1,width=0.1,label='Adventure')
ax.bar(x-0.3,y2,width=0.1,label='fighting')
ax.bar(x-0.2,y3,width=0.1,label='misc')
ax.bar(x-0.1,y4,width=0.1,label='platform')
ax.bar(x,y5,width=0.1,label='puzzle')
ax.bar(x+0.1,y6,width=0.1,label='racing')
ax.bar(x+0.2,y7,width=0.1,label='role-playing')
ax.bar(x+0.3,y8,width=0.1,label='shooter')
ax.bar(x+0.4,y9,width=0.1,label='simulation')
ax.bar(x+0.5,y10,width=0.1,label='sports')
ax.bar(x+0.6,y11,width=0.1,label='strategy')

ax.set_xticks(x)
ax.set_xticklabels(rata2.index)
ax.set_title('Mean Sales by video game genre')
ax.set_ylabel('mean sales')
ax.set_xlabel('Region sales')


ax.legend(loc='upper right')
plt.style.use('ggplot')

plt.show()
```

<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>
