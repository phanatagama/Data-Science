```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True, sharex=True, figsize=(18,8))
ax[0].plot(x,tan, color='b', label='tanx',)
ax[1].plot(x,sin, color='r', label='sinx')
ax[2].plot(x,cos, color='g', label='cosx')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik Sinx')
ax[2].set_title('Grafik Cosx')

fig.legend(loc='upper center')

plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(14.2,9), )
ax.plot(x,y)
ax.set_title('Grafik Cosx')

col1 = ax.inset_axes([-0.1 , 0.9 , 0.3 , 0.3])
col1.plot(x,y5, color='y')
col1.plot(x,y, color='b')
col1.set_title('Grafik cos^5x')

col2 = ax.inset_axes([-0.1 , -0.1 , 0.3 , 0.3])
col2.plot(x,y2, color='r')
col2.plot(x,y, color='b')
col2.set_title('Grafik cos^2x')

col3 = ax.inset_axes([0.9 , 0.9 , 0.3 , 0.3])
col3.plot(x,y4, color='black')
col3.plot(x,y, color='b')
col3.set_title('Grafik cos^4x')

col4 = ax.inset_axes([0.9 , -0.1 , 0.3 , 0.3])
col4.plot(x,y3, color='g')
col4.plot(x,y, color='b')
col4.set_title('Grafik cos^3x')



plt.show()
```
